```python
test_live = []
for img in glob.glob("/content/drive/My Drive/Spoof_data/Testing Biometrika Live/live/*.png"):
    print(img)
    n= cv2.imread(img) #reading the testing live data
    test_live.append(n.flatten()) #flattening the data to convert into 1D vector for testing
```

```python
test_spoof = []
for img in glob.glob("/content/drive/My Drive/Spoof_data/Testing Biometrika Spoof/Testing Biometrika Spoof/spoof/*.png"):
    print(img)
    n= cv2.imread(img) #reading the testing spoof data
    test_spoof.append(n.flatten()) #flattening the data to convert into 1D vector for testing
```

```python
test_spoof
```

```python
finger_print_X = []
c=0
for image in  train_spoof:
  finger_print_X.append(image) 
  c+=1
print(c)  
finger_print_X = finger_print_X[1:155] #slicing image at 155th position 
for image in train_live:
  finger_print_X.append(image) #append the image into the array 
  
```

```python
finger_print_Y = []
for image in train_spoof:
  finger_print_Y.append('spoof')
finger_print_Y = finger_print_Y[1:155] #slicing images
for image in train_live:
  finger_print_Y.append('live') #append the image into the array
```

```python
X_train = np.array(finger_print_X) 
Y_train = np.array(finger_print_Y)
```

```python
finger_print_X = []
for image in test_live:
  finger_print_X.append(image)
for image in  test_spoof:
  finger_print_X.append(image)
```

```python
finger_print_Y = []
for image in test_live:
  finger_print_Y.append('live')
for image in test_spoof:
  finger_print_Y.append('spoof')
```

```python
X_test = np.array(finger_print_X)
Y_test = np.array(finger_print_Y)
```

```python
linear = svm.SVC(kernel='linear', C=1, decision_function_shape='ovo').fit(X_train, Y_train)
```

```python
Y_pred=linear.predict(X_test)
```

```python
print("FingerPrint Spoof Detection Parameters")
print("Overall Accuracy using Linear SVM: "+str(accuracy_score(Y_test, Y_pred)))
#print('\n')
print(classification_report(Y_test, Y_pred))
cm = confusion_matrix(Y_test, Y_pred)
print("Accuracy for each class")
print(cm.diagonal()/cm.sum(axis=1))
```
